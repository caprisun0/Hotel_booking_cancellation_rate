# 데이터 분석 프로젝트 배경 및 목표 탐색

## 프로젝트 배경
- 시나리오  
1. 최근 모기업에서 조직의 체질개선을 위해 호텔사업을 정리하려고 한다. 현재 모기업에서 운영 중인 호텔은 (Resort Hotel, City Hotel) 두 곳인데, 모기업은 이 중 상대적으로 경제성이 낮은 쪽을 정리하길 원한다. 주어진 데이터셋을 분석하여 각 호텔의 경제성을 평가한 후 의견을 제시하라.
2. 당신은 호텔사업을 운영하는 기업의 데이터팀 소속 분석가이다. 최근 기업에서는 호텔 사업 확장을 위해 새로운 호텔을 보유하였다. 기업은 당신에게 기존에 운영 중인 호텔 두 곳의 데이터셋을 분석하여 보고서를 작성하길 요구했다. 각 호텔의 강점과 약점을 평가하고 이를 근거로 새로운 호텔에 적용할 사업 전략을 보고서로 제시하라.
3. 당신이 소속된 기업에서 최근 두 호텔을 예약한 고객 중 Cancel 및 No-Show 고객이 증가하여 이윤이 저하되고 있다. 이를 막고자 기업은 Cancel 및 No-Show 고객을 예측하기 위한 프로젝트를 구상 중이며, 사전에 당신에게 두 유형의 고객에 대한 특성을 파악하도록 요구했다. 두 호텔의 Cancel 및 No-Show 고객의 현황 및 특성, 성향을 파악하여 보고서로 작성하라.
4. 기업에서는 호텔 사업 확장을 위해 공격적인 마케팅을 시도하고자 한다. 이를 위해 호텔을 이용하는 고객 세그먼트를 먼저 분류할 것을 당신에게 요구했다. 데이터셋을 기반으로 두 호텔의 고객 세그먼트를 분류하고 그 근거를 제시하며, 각 고객 세그먼트 별 접근 전략을 기획하라.
5. 당신이 소속된 팀의 팀장이 두 호텔의 객단가(ADR)가 높은 고객군의 특성을 알고 싶어한다. 객단가에 영향을 미치는 요인을 분석해 보고서를 작성하라.
6. 당신이 소속된 팀의 팀장이 호텔과 관계된 여행사 별 실적을 알고 싶어한다. 여행사별 분석 보고서를 작성하라.
7. 당신이 소속된 팀의 팀장이 당신에게 비용 절감을 위한 방안을 제시하도록 요구했다. 호텔의 여러 서비스 중에서 경제성이 떨어지는 요소를 분석하여 호텔의 운영 비용 절감 방안을 제시하라.

## 데이터 설명

- 이 데이터 셋은 두 호텔의 수요(예약) 데이터를 담고 있다. 
- 한 호텔은 리조트 호텔이며 다른 호텔은 시티 호텔이다.
- 데이터셋에는 31개의 변수가 담겨있다. 
- 2015년 7월 1일부터 2017년 8월 31일까지의 기간 내의 데이터이다.
- 데이터는 각 호텔의 PMS 데이터베이스에서 직접 SQL로 추출되었다.
- 두 호텔은 모두 포르투갈에 있다. 리조트 호텔은 알가르베에 위치하며, 시티 호텔은 리스본에 위치한다.
- 각 필드의 값은 가능하면 최종적인 값이 아니라 입실 전날의 예약 변경 로그 데이터베이스(BL)에서 추출하고자 했다.
- (예측 모델을 개발할 때 이미 관측된 결과값이 예측 모델에 삽입되는 것을 방지하기 위해서.)   
- 해당 예약의 예약 변경 로그가 존재하지 않으면 예약 정보 데이터베이스(BO)의 값을 추출했다.
- meals, distribution channels, nationalities or market segments의 값은 관련 데이터베이스에서 추출했다.
- PMS 데이터베이스에는 결측치가 존재하지 않음을 확인했다. 그러므로 NULL 값은 결측치가 아니라 '해당 없음'이다.
- 대신 Country 같은 경우는 입실 전까지 알 수 없는 경우도 많다. 이런 경우는 결측치라고 볼 수 있다.
- 입실할 때나 숙박 중에도 예약 시 등록한 인원수, 숙박 기간, 객실 유형 등이 계속 변경될 수 있다.

### City Hotel과 Resort Hotel

City Hotel: 도시에 위치한 호텔로, 주로 도심이나 번화가에 위치합니다. 비즈니스 여행객이나 관광객들이 편리하게 접근할 수 있도록 대도시의 중심부에 위치합니다. 교통이 편리하며 주변에 다양한 상업 시설과 엔터테인먼트 시설이 위치해 있습니다. 주로 비즈니스 관련 시설과 서비스를 제공하며, 대부분의 고객은 비즈니스 관련 목적을 위해 방문합니다.   
Resort Hotel: 휴양지나 관광지에 위치한 호텔로, 자연이 아름다운 지역이나 해변가에 자리하고 있습니다. 휴양과 휴식을 위한 특별한 환경을 제공하며, 주로 가족 여행객이나 휴가를 즐기기 위한 관광객들이 방문합니다. 주변에는 자연 경관이 풍부하며, 특별한 레저 시설이나 액티비티를 즐길 수 있는 장소들이 있습니다. 대부분의 Resort Hotel은 호텔 내 레스토랑, 수영장, 스파, 골프장 등 다양한 편의 시설을 갖추고 있습니다. 

### PMS   
- https://www.oracle.com/kr/hospitality/what-is-hotel-pms/ (Oracle)     
- 호텔 자산 관리 시스템(PMS)이란 무엇인가?
호텔 자산 관리 시스템(PMS)은 우수한 고객 경험 제공을 포함한 호텔 비즈니스 운영의 모든 측면을 관리합니다. 과거 호텔 PMS는 하나의 호텔 또는 호텔 그룹이 객실 예약, 고객 체크인/체크아웃, 객실 배정, 객실 요율 관리, 청구 등 프론트 오피스 기능들을 관리할 수 있게 해 주는 소프트웨어를 의미했습니다. 호텔 PMS는 시간 소모적인, 서류 집약적인 프로세스를 대체했죠. 하지만 오늘날 호텔 PMS 기술은 프론트 데스크 너머로 크게 확장되었습니다. 이제 호텔 PMS는 호텔 경영자가 탁월한 고객 경험을 제공할 수 있게 지원하는 핵심적인 비즈니스 운영 시스템이 되었습니다.

### CRS와 GDS
- 여행신문(https://www.traveltimes.co.kr)
- CRS는 컴퓨터예약시스템(Computerized Reservations System)의 약자입니다. 미국 아메리칸 항공에서 1964년 만든 세이버(SABRE)가 그 시초이며, 당시에는 항공업무자동화를 위해 항공사 측에서 개발한 전산예약시스템이었으나 최근에는 항공 뿐 아니라 호텔 및 렌터카, 정보제공 등으로 영역을 확대해 ‘여행에 관한 종합적인 서비스를 제공하는 전산예약시스템’을 의미하게 됐습니다.  
GDS(Global Distribution System)는 CRS가 발전해 나가는 과정에서 탄생됐습니다. 세이버가 각 여행사 대리점들에 설치되면서 업무자동화와 항공권 판매 증진이라는 막대한 효과를 가져왔고, 이에 따라 각 항공사들마다 경쟁적으로 CRS를 개발하기 시작했습니다. 유럽 및 아시아에서도 여러 항공사들이 연합해 새로운 CRS를 개발했는데요. 그렇게 만들어진 것이 BR, SR, AZ, KL 등 유럽항공사들이 만든 아마데우스, CX, SQ, AZ, KL 등 동남아시아 기반의 항공사들이 연합한 애바카스 등입니다. 즉 GDS는 ‘CRS가 글로벌하게 진화한 형태’로 보시면 되겠습니다.

```python
# 3, 4, 5, 6번 필드로 입실 날짜(arrival_date)를 조합할 수 있음.
# 먼저 arrival_date_month를 숫자로 변경해야 함.
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}
# 딕셔너리와 일치하지 않으면 NaN을 반환하지 않도록 조건 추가
df['arrival_date_month'] = df['arrival_date_month'].apply(lambda x: month_mapping[x] if x in month_mapping else x)
df['arrival_date_month'] = pd.to_numeric(df['arrival_date_month'], errors='coerce')

# 입실 날짜 변환
df['arrival_date'] = pd.to_datetime({
    'year': df['arrival_date_year'],
    'month': df['arrival_date_month'],
    'day': df['arrival_date_day_of_month']
})
df.head()
```

```python
# arrival_date를 만들었으므로, 기존에 있던 'arrival_date_year','arrival_date_month', 'arrival_date_day_of_month' 제거
# 'arrival_date_week_number' 필드는 주차별 분석을 위해 남겨둠.
fields_to_remove = ['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']
df = df.drop(columns=[col for col in fields_to_remove if col in df.columns])
df.head()
```

```python
# lead_time(PMS 입력 날짜로부터 입실 날짜까지 경과한 일 수) 및 arrival_date로 예약을 한 날짜(reservaition_date)를 구할 수 있다.
df['reservation_date'] = df['arrival_date'] - pd.to_timedelta(df['lead_time'], unit='D')
df.loc[:,['arrival_date','lead_time','reservation_date']]
```

```python
# stays_in_weekend_nights, stays_in_week_nights로 해당 예약이 몇 박인지 계산할 수 있다.
# df['stays_in_nights']
df['stays_in_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
# 투숙객의 체크아웃 날짜('departure_date') 계산
df['departure_date'] = df['arrival_date'] + pd.to_timedelta(df['stays_in_nights'], unit='D')
```
